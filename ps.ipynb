```python
from google_play_scraper import search, app
import pandas as pd
import re

# List of keywords for which to perform searches
keywords = [
    "Cybersecurity",
    "Cyber Security",
    "Information Security",
    "IT Security",
    "System Security",
    "Network Security",
    "Mobile Security"
]

# Create a Pandas Excel writer using the openpyxl engine
excel_filename = 'app_details.xlsx'
with pd.ExcelWriter(excel_filename, engine='openpyxl') as writer:
    for keyword in keywords:
        # Perform the search
        search_results = search(
            keyword,
            lang="en",
            country="us",
            n_hits=30
        )

        # Prepare data for the DataFrame
        data = []
        for search_result in search_results:
            app_id = search_result['appId']
            try:
                # Fetch full details of the app
                details = app(app_id)
                app_name = details['title']
                price = details['price']
                installs = details['realInstalls']
                ratings = details['score'] if 'score' in details else "0"
                # Append to data list
                data.append([app_id, app_name, price, installs, ratings])
            except Exception as e:
                print(f"An error occurred while fetching details for app {app_id}: {e}")

        # Create a DataFrame
        df = pd.DataFrame(data, columns=["App ID", "App Name", "Price", "Installs", "Ratings"])

        # Write DataFrame to a sheet with the name of the keyword
        safe_keyword = re.sub(r"[^\w\s]", '', keyword).replace(" ", "_").lower()
        df.to_excel(writer, sheet_name=safe_keyword, index=False)

    print(f"Excel file '{excel_filename}' has been created with separate sheets for each keyword.")
```
